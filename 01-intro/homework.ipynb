### Q4. One-hot encoding

| Orient      | Description                           | Example result                                |
| ----------- | ------------------------------------- | --------------------------------------------- |
| `'records'` | List of row dictionaries              | `[{col: val, ...}, ...]`                      |
| `'dict'`    | Column-wise dict of lists             | `{col: [val1, val2, ...]}`                    |
| `'list'`    | Dict of columns with values as lists  | Same as `'dict'`                              |
| `'series'`  | Dict of columns with values as Series | `{col: pd.Series(...)}`                       |
| `'split'`   | Dict with keys: index, columns, data  | `{index: [...], columns: [...], data: [...]}` |
| `'index'`   | Dict of dicts keyed by row index      | `{index: {col: val}}`                         |


```python
categorical = ['PULocationID', 'DOLocationID']
df_jan_dropped[categorical] = df_jan_dropped[categorical].astype(str)
```

```python
train_dicts = df_jan_dropped[categorical].to_dict(orient='records') # One-hot encoding of categorical features.
```

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
```

```python
X_train.shape
```

### Q5. Training a model

```python
target = 'duration'
y_train = df_jan_dropped[target].values

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_train)

#rms = mean_squared_error(y_train, y_pred, squared=False)
root_mean_squared_error(y_train, y_pred)
```

```python
sns.distplot(y_pred, label='prediction')
sns.distplot(y_train, label='actual')

plt.legend()
```

### Q6. Evaluating the model

```python
def read_data(filename,df):
     if filename.endswith('.parquet'):
        df = pd.read_parquet(filename)
        df['duration']= df['tpep_dropoff_datetime']-df['tpep_pickup_datetime']
        df['duration'] = df['duration'].apply(lambda td: td.total_seconds()/60)
        df = df[(df['duration'] >= 1) & (df['duration'] <= 60)]
        categorical = ['PULocationID', 'DOLocationID']
        df[categorical] = df[categorical].astype(str)
        print(filename, '\n', df.shape)
        return df
```

```python
cat_cols = ['PULocationID', 'DOLocationID']
df_jan = read_data('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet', cat_cols)
```

```python
df_feb = read_data('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet', cat_cols)
```

```python
dv = DictVectorizer()

train_dicts = df_jan[cat_cols].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_feb[cat_cols].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
target = 'duration'
y_train = df_jan[target].values
y_val = df_feb[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)

root_mean_squared_error(y_val, y_pred)
```
